```python
# python의 예외처리

def my_func(list_data):
    
    my_sum = 0    
    try:
        my_sum = list_data[0] + list_data[1] + list_data[2]
        
    except Exception as err:
        print('실행시 문제가 발생했어요!!')
        my_sum = 0
    
    else:
        print('실행시 문제가 없어요!!')
        
    finally:
        print('만약 finally가 존재하면 무조건 실행되요!!')
                
    return my_sum
    

my_list = [1,2]    
print(my_func(my_list))
```

```python
# 기본적인 file 처리에 대해서 알아보아요!!
# file처리는 대부분 pandas를 이용해서 처리합니다.!

my_file = open('mpg.txt','r')

# 파일안에 있는 모든 내용(line)을 화면에 출력할 꺼예요!
# ''  공백문자열은 False로 간주
while True:
    line = my_file.readline()
    print(line)
    if not line:
        break

my_file.close()   # 반드시 사용한 resouce는 적절하게 해제처리를 해 줘야 해요!     
```
