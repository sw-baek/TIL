```python
arr = np.array([1.2,2.3,3.5,4.1,5.7])
print(arr)

arr = arr.astype(np.int32)
print(arr)

arr_2 = np.zeros((3,4), dtype = np.int32)
print(arr_2)

arr_3 = np.ones((3,4), dtype=np.float64)
print(arr_3)

arr_4 = np.empty((3,4))
print(arr_4)

arr = np.full((3,4), 7, dtype = np.float64)
print(arr)
```
