```python
import pandas as pd
movies = pd.read_csv("C:/python_ML/movies.csv")
rating = pd.read_csv("C:/python_ML/ratings.csv")

movies
rating
#1
print(rating['rating'].mean())


#2
rating.groupby(rating['userId'])['rating'].mean()

#3 
display(movies)
ratings=rating.groupby(rating['movieId'])['rating'].mean()

df = pd.merge(movies, ratings, left_on="movieId", right_index=True, how= "inner")

```

```python
#4
df_2 = df.loc[:,['movieId','title','rating']]
df.loc[df["rating"]==df["rating"].max(),["title","rating"]]
```

```python
#5
df_2=df.loc[df['genres'].str.contains("Comedy")].loc[df["rating"]==df["rating"].min()]
df_2.sort_values(by = 'title')
```

```python
# 6 번
import time
#영화아이디 기준으로 판다스 합병
new_df = pd.merge (movies, rating, left_on = "movieId", right_on = "movieId", how = "inner")
#2015년 1월 1일이 컴퓨터 시점으로 언제인지 찾기
start = time.mktime(time.strptime("2015-01-01 00:00:00","%Y-%m-%d %H:%M:%S"))
#2015년 언제 끝나는지 컴퓨터 시점으로 찾기
end = time.mktime(time.strptime("2016-01-01 00:00:00","%Y-%m-%d %H:%M:%S"))
#로맨스 장르인 것들만 데이터 프레임에서 찾기
new_df_2=new_df.loc[new_df["genres"].str.contains("Romance")]
#2015년에 해당하는 것만 찾기
new_df_3=new_df_2.loc[(new_df_2["timestamp"] > start)&(new_df_2["timestamp"] < end)]
#점수 평균 내기
new_df_3["rating"].mean()
```
