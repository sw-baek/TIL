```python
import numpy as np
import pandas as pd

my_dict = {'학과' : ['컴퓨터','경영학과','컴퓨터','경영학과','컴퓨터'],
          '학년' : [1,2,3,2,3],
          '이름' : ['홍길동','신사임당','김연아','아이유','강감찬'],
          '학점' : [1.5,4.4,3.7,4.5,4.2]}

df = pd.DataFrame(my_dict)
display(df)

#학과를 기준으로 grouping ~
# score = df[['학점','이름']].groupby(df['학과'])
# print(score)

#그룹안에 데이터를 확인하고 싶은 경우에는 get_group()

# print(score.get_group('경영학과'))

# print(score.size())
# print(score.mean())

dept = df['학점'].groupby([df['학과'],df['학년']]) # fancy indexing 으로 해줘야
display(dept)
print(dept.mean())
dept.mean().unstack()  # 최하위 index를 column으로 변경 
```

```python
my_dict = {'학과' : ['컴퓨터','경영학과','컴퓨터','경영학과','컴퓨터'],
          '학년' : [1,2,3,2,3],
          '이름' : ['홍길동','신사임당','김연아','아이유','강감찬'],
          '학점' : [1.5,4.4,3.7,4.5,4.2]}

df = pd.DataFrame(my_dict)
display(df)

score = df.groupby(df['학과'])
print(score.get_group('경영학과'))
print(score.size())
print(score.mean())
```
